```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("adult.csv")
data.head()
```

```python
data.columns=["age","workclass","fnlwgt","education","education-num","marital-status","occupation","relationship","race","sex","capital-l","capital-g","hpw","native-country","income"]
```

```python
#histogram of all the number data
numeric_col = ["age","fnlwgt","education-num","capital-l","capital-g","hpw"]
numeric_data= data[numeric_col]
```

```python
sns.histplot(data=data,x="age", bins=20,kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
numeric_data.hist(bins=20, figsize=(10,6), edgecolor="black")
plt.tight_layout()
```

```python
#dot plots 
sns.stripplot(data=data,x="income",y="education-num",jitter=True)
```

```python
sns.countplot(data=data, x="sex")
```

```python
sns.lineplot(data=data,x="age",y="hpw", marker="o")
```

```python
sns.jointplot(data=data,x="age",y="hpw")
```

```python
data["income"].value_counts().plot.pie( autopct="%1.1f%%")
```

```python
sns.boxplot(data=data,x="income",y="hpw", hue="income")
```
