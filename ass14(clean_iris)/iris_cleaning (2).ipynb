```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("iris.csv")
data.head()
```

```python
data.columns = [ 'sepal_length', 'sepal_width', 'petal_length', 'petal_width','species']
data
```

```python
data.replace("?",np.nan,inplace=True)
```

```python
data.dropna()
```

```python
numeric_data =["sepal_length","sepal_width","petal_length","petal_width"]
```

```python
for col in numeric_data:
    data[col]=pd.to_numeric(data[col],errors="coerce")
```

```python
data = data[(data[numeric_data] >= 0).all(axis=1)]
```

```python
Q1 = data[numeric_data].quantile(0.25)
Q3 = data[numeric_data].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
higher = Q3 + 1.5 * IQR

# Remove outliers
data_clean = data[
    ~((data[numeric_data] < lower) | 
      (data[numeric_data] > higher)).any(axis=1)
]

print(data_clean)
```

```python
le = LabelEncoder()

data["species"] = le.fit_transform(data["species"])
```
