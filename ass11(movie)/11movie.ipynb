```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("movies_metadata.csv", sep=",")
data.head()
```

```python
data["original_language"].unique()
```

```python
english = data[data['original_language']=='en']
french = data[data['original_language']=='fr']
```

```python
merged_data = pd.concat([english,french])
print(merged_data.head())
```

```python
sorted_data = data.sort_values(
    by='vote_average',
    ascending=False)

print(sorted_data)
```

```python
transpose = sorted_data.transpose()
print(transpose)
```

```python
data['id']= data.index
melted_data= pd.melt(data, id_vars=['id','title'])
print(melted_data)
```

```python
wide_data = melted_data.pivot(index=['id','title'], columns='variable', values='value')
```
