```python
data.replace("?",np.nan,inplace=True)
```

```python
numeric_data = data.select_dtypes(include='number')
print(numeric_data)
```

```python
data=data[(numeric_data>=0).all(axis=1)]
```

```python
data.dropna()
```

```python
#outliers

Q1=numeric_data.quantile(0.25) 
Q3=numeric_data.quantile(0.75)

IQR= Q3-Q1

lower= Q1-1.5*IQR
higher=Q3+1.5*IQR

data_clean= data[~((numeric_data<lower) | (numeric_data>higher)).all(axis=1)]

```

```python
print(data_clean)
```
