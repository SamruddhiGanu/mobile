```python
import requests
import pandas as pd
```

```python
url="https://store.steampowered.com/appreviews/730?json=1"
```

```python
response = requests.get(url)
```

```python
data = response.json()
```

```python
reviews_data=[]
```

```python
reviews=data['reviews']
```

```python
for review in reviews:
    cust_name = review['author']['personaname']
    comment=review['review']
    ratings ="recommended" if review['voted_up'] else "not recomm"

    reviews_data.append([cust_name, comment,ratings])
```

```python
df =pd.DataFrame(reviews_data,columns=[ 'Customer Name', 'Comment', 'Rating'])
```

```python
print(df)
```

```python
df.to_csv('steam_reviews.csv', index=False)
print("\nCSV File Saved Successfully")
```
