```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Iris.csv", sep=",")
```

```python
print(data.head())
```

```python
print(data.columns)
```

```python
data.columns = [ 'sepal_length', 'sepal_width', 'petal_length', 'petal_width','species']
```

```python
print(data.columns)
```

```python
setosa = data[data['species']=='Iris-setosa']
```

```python
versicolor = data[data['species']=='Iris-versicolor']
virginica = data[data['species']=='Iris-virginica']
```

```python
print(setosa)
```

```python
merged_subset = pd.concat([versicolor, virginica],ignore_index=True)
print(merged_subset)
```

```python
sorted_data = data.sort_values(
    by="petal_length",
    ascending =False )
    
```

```python
print(sorted_data)
```

```python
transpose = sorted_data.transpose()
print(transpose)
```

```python
data["id"] = data.index
melted_data = pd.melt(data,id_vars=['id','species'], var_name="parameter",value_vars = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
print(melted_data)
```

```python
print(data.shape)
```

```python
wide_data = melted_data.pivot(index=['id','species'],columns="parameter",values="value")
```

```python
print(wide_data)
```
