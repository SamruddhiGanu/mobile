```python
import pandas as pd
import numpy as np
```

```python
data  = pd.read_csv('dataset_Facebook.csv',sep=';')
```

```python
data.head()
```

```python
photo_posts = data[data['Type'] == 'Photo']
status_posts = data[data['Type'] == 'Status']
```

```python
print(photo_posts.head())
```

```python
merged_data = pd.concat([photo_posts, status_posts], ignore_index=True)
```

```python
print(merged_data.head())
```

```python
print(merged_data)
```

```python
video_post = data[data['Type']=='Link']
```

```python
print(video_post.head())
```

```python
sorted_data = data.sort_values(
    by='Page total likes',
    ascending=False
)
```

```python
print("\nSorted Data")
print(sorted_data[['Type', 'Page total likes']])
```

```python
transposed_data = sorted_data.transpose()
```

```python
print(transposed_data)
```

```python
melted_data = pd.melt(data,
                      id_vars=['Type', 'Category', 'Post Month', 'Post Weekday'],
                      value_vars=['like', 'share', 'comment'],
                      var_name='Interaction Type',
                      value_name='Interaction Count')
```

```python
print(melted_data)
```

```python
wide_data = melted_data.pivot_table(index=['Type', 'Category', 'Post Month', 'Post Weekday'],
                                    columns='Interaction Type',
                                    values='Interaction Count',
                                    aggfunc='sum').reset_index()
```
